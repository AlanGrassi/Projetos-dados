```python
#Importação de pacotes
import numpy as np
import pandas as pd
```

```python
# Importa base de dados
df=pd.read_csv('agricultural_raw_material pt.csv')
```

```python
df.info
#Checagem dor valores nulos de cada coluna
df.isnull().sum()
```

Lidar com dados ausentes, incorretos e inválidos

```python

df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

df = df.dropna()

df.isnull().sum()

lst = ["Preço da lã grossa", "% de variação do preço da lã grossa", "Preço do copra",
       "% de variação do preço do copra","Preço do algodão","% de variação do preço do algodão",
       "Preço da lã fina", "% de variação do preço da lã fina", "Preço da tora dura",
       "% de variação do preço da tora dura", "Preço da madeira serrada dura",
       "% de variação do preço da madeira serrada dura", "Preço do couro", "% de variação do preço do couro",
       "Preço do compensado", "% de variação do preço do compensado", "Preço da borracha",
       "% de variação do preço da borracha", "Preço da tora macia", "% de variação do preço da tora macia",
       "Preço da madeira serrada macia", "% de variação do preço da madeira serrada macia",
       "Preço da polpa de madeira", "% de variação do preço da polpa de madeira"]
df[lst] = df[lst].astype("float")
df.dtypes
```

# Colunas de data e hora colunas

Formatando a coluna datetime e definindo-a como índice para o conjunto de dados

```python
df['Mês']  = pd.to_datetime(df['Mês'].str.upper(), format='%b%y', yearfirst=False)
# Mês de referência
df = df.set_index('Mês')
```

```python
df.head()
```

Análise Exploratória e Visualização Vamos começar importando matplotlib.pyplot e seaborn.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Mapa de Calor


```python
raw_data=["Preço da lã grossa", "Preço do copra","Preço do algodão", "Preço da lã fina", "Preço da tora dura", "Preço da madeira serrada dura", "Preço do couro", "Preço do compensado", "Preço da borracha", "Preço da tora macia", "Preço da madeira serrada macia", "Preço da polpa de madeira"]
# obtendo a matriz de correlação
corrmat = df[raw_data].corr()
# definindo o tamanho do gráfico
fig = plt.figure(figsize = (14, 10))
# ocultando a parte superior da matriz, pois ela é simétrica (repetitiva)
mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax = .8,mask=mask, square = True, annot = True)
plt.show()
```

O mapa de calor mostra a correlação entre as matérias-primas maior o valor correlacionado maior chance de serem duas matérias-primas relacionadas, mas não necessariamente

##Fazendo e respondendo perguntas
Q1: Descubra a variação normal do preço de cada matéria-prima

```python
changelist=["% de variação do preço da lã grossa", "% de variação do preço do copra", "% de variação do preço do algodão", "% de variação do preço da lã fina", "% de variação do preço da tora dura", "% de variação do preço da madeira serrada dura", "% de variação do preço do couro", "% de variação do preço do compensado", "% de variação do preço da borracha", "% de variação do preço da tora macia", "% de variação do preço da madeira serrada macia", "% de variação do preço da polpa de madeira"]
for i in range(len(changelist)):
    plt.figure(figsize=(12,12))
    df[changelist[i]].hist(figsize=(11, 9), linewidth=1)
    plt.xlabel('Variação %')
    plt.ylabel('contagem')
    plt.legend(changelist[i:],loc='upper center',bbox_to_anchor=(1.2,1))
```

Podemos observar que a maioria das matérias-primas tem % de mudança frequente ideal inferior a 5%

#Q2: Encontre a matéria-prima que tem o menor preço ao longo dos anos

```python
plt.figure(figsize=(12, 12))
materialslist=["Preço da lã grossa","Preço do copra","Preço do algodão","Preço da lã fina","Preço da tora dura", "Preço da madeira serrada dura", "Preço do couro", "Preço do compensado", "Preço da borracha", "Preço da tora macia", "Preço da madeira serrada macia", "Preço da polpa de madeira"]
for i in range(len(materialslist)):
    plt.subplot(4,3,i+1)
    plt.subplots_adjust( hspace=1 ,wspace=0.5)
    plt.title(materialslist[i])
    plt.plot(df[materialslist[i]])
    plt.xticks(rotation=90)
plt.suptitle("Comparação de Preços de Matérias-Primas")
```

Podemos ver que o algodão e a borracha são de preços mais baixos vamos comparar preços para entender melhor qual é o mais baixo

```python
plt.figure(figsize=(10, 10))
plt.plot(df[["Preço do algodão", "Preço da borracha"]])
plt.title("Comparação de Preço de Matérias-Primas")
plt.xlabel("Anos")
plt.ylabel('Preços')
plt.legend(["Preço do algodão", "Preço da borracha"],loc='upper center',bbox_to_anchor=(1.2,1))
```

Algodão é matéria-prima de menor preço nos últimos anos

Tanto o algodão quanto a borracha são materiais de baixo preço

```python
plt.figure(figsize=(10, 10))
plt.plot(df[["Preço da lã grossa","Preço do copra","Preço do algodão","Preço da lã fina","Preço da tora dura", "Preço da madeira serrada dura", "Preço do couro", "Preço do compensado", "Preço da borracha", "Preço da tora macia", "Preço da madeira serrada macia", "Preço da polpa de madeira"]])
plt.title("Comparação de Preços das Matérias-Primas")
plt.xlabel('Anos')
plt.ylabel('Preços')
plt.legend(["Preço da lã grossa","Preço do copra","Preço do algodão","Preço da lã fina","Preço da tora dura", "Preço da madeira serrada dura", "Preço do couro", "Preço do compensado", "Preço da borracha", "Preço da tora macia", "Preço da madeira serrada macia", "Preço da polpa de madeira"],loc='upper center',bbox_to_anchor=(1.2,1))
```

A partir dos gráficos pudemos analisar as matérias-primas em tipos de acordo com seu preço ao longo dos anos

#Q3: qual matéria-prima tem a maior e menor variação de % de preço

```python
import random as random
```

```python
plt.figure(figsize=(22,22))
for i in range(len(changelist)):

    r = random.random()
    b = random.random()
    g = random.random()
    color = (r, g, b)
    plt.subplot(4,3,i+1)
    plt.subplots_adjust( hspace=1 ,wspace=0.5)
    plt.plot(df[changelist[i]],c=color)
    plt.xticks(rotation=90)
    plt.title(changelist[i])
    plt.xlabel('Anos')
    plt.ylabel('Varianção %')
```

Q4: Encontre as matérias-primas com mudança drástica de preço

```python
lowlist=["Preço do algodão", "Preço do couro", "Preço da tora macia", "Preço da tora dura", "Preço da madeira serrada macia", "Preço da borracha"]

plt.figure(figsize=(12,12))
plt.ylabel('Prices')
plt.xlabel('Years')

for i in range(len(lowlist)):
    sns.scatterplot(y=df[lowlist[i]],x=df.index);
    plt.legend(lowlist,loc='upper center',bbox_to_anchor=(1.2,1))
```

A mudança de preço é drástica para o preço de tora dura entre materiais de baixa faixa de preço e entre materiais de alto preço, são os preços de lã fina
